```python
import numpy as np
import matplotlib.pyplot as plt
from lapjv import lapjv
from scipy.spatial.distance import cdist
```

```python
embedding = np.load("data/TGGASmallUmap.npy")
```

```python
embedding.shape
```

```python
length = embedding.shape[0]
labels = np.random.choice([0, 1, 2, 3], size=length)
colours = ['b', 'k', 'g', 'r']
```

```python
plt.scatter(embedding[:, 0], embedding[:, 1], c=labels);
```

```python
size1 = int(np.ceil(np.sqrt(length)))
size2 = int(np.ceil(length/size1))
grid_size = (size1, size2)

grid = np.dstack(np.meshgrid(np.linspace(0, 1, size2), 
                             np.linspace(0, 1, size1))).reshape(-1, 2)
grid_map = grid[:length]  # non-rectangular grid
cost_matrix = cdist(grid_map, embedding, "sqeuclidean")
cost_matrix = cost_matrix * (100000 / cost_matrix.max())
row_asses, col_asses, _ = lapjv(cost_matrix)

print(grid_size, grid.shape, grid_map.shape, len(row_asses), len(col_asses))
```

```python
grid_jv = grid_map[col_asses]

for i, end in enumerate(grid_jv):
    plt.plot(end[0], end[1], color=colours[labels[i]], marker='o');
```
